```python
import networkx as nx
import random
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from tqdm import tqdm 
```

```python
probabilities= []
for i in range(0, 11, 2):
  probabilities.append(i/10)
print(probabilities)
numedges= []
for j in range(1, 1000, 200):
  numedges.append(j)
print(numedges)

numedges.append(1000)
```

```python
df = pd.DataFrame()
df2 = pd.DataFrame()
list = []
```

```python

for i in probabilities:
  for j in numedges:
    dict= {}
    G = nx.powerlaw_cluster_graph(1000, j, i)
    close = nx.closeness_centrality(G)
    AND = nx.average_neighbor_degree(G)
    clust = nx.average_clustering(G)
    dict["Type"] = "HK"
    dict["Probability"] = i
    dict["Num_Edges"] = j
    dict["closeness"] = close
    dict["avg_neighbor_degree"] = AND
    dict["avg_clustering"] = clust
    df = df.append(dict, ignore_index=True)
    dict= {}
    degree_sequence = [d for n, d in G.degree()] 
    G_CONFIG= nx.configuration_model(degree_sequence)
    G_CONFIG=nx.Graph(G_CONFIG)
    close_CONFIG = nx.closeness_centrality(G_CONFIG)
    AND_CONFIG = nx.average_neighbor_degree(G_CONFIG)
    clust_CONFIG = nx.average_clustering(G_CONFIG)
    dict["Type"] = "CONFIG"
    dict["Probability"] = i
    dict["Num_Edges"] = j
    dict["closeness"] = close_CONFIG
    dict["avg_neighbor_degree"] = AND_CONFIG
    dict["avg_clustering"] = clust_CONFIG
    df2 = df2.append(dict, ignore_index=True)
```
